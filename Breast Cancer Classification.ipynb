```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import (
    confusion_matrix,
    ConfusionMatrixDisplay,
    accuracy_score,
    classification_report,
    precision_score,
    recall_score,
    f1_score
)
```

```python
params = {
    'font.size': 18,
    'font.family': 'Times New Roman',
    'figure.figsize': (8, 2),
    'figure.dpi': 600,
    'xtick.labelsize': 18,
    'ytick.labelsize': 18
}
plt.rcParams.update(params)
```

# Breast Cancer Wisconsin Dataset

```python
# Load data
Data = load_breast_cancer()

# Two classes
print(Data.target_names)

# Data + label
X, y = Data.data, Data.target

# Two features
F1 = X[:,4:5]
F2 = X[:,21:22]
print("\nTwo features:")
print(Data.feature_names[4])
print(Data.feature_names[21])

# Dataset contains only two features
X_reduced = np.hstack((F1,F2))

# Training set + test set
X_train, X_test, y_train, y_test = train_test_split(X_reduced, y, test_size = 0.3, random_state = 8)
```

# Model Training & Prediction

```python
gnb = GaussianNB()

# Model Training
gnb.fit(X_train, y_train)

# Prediction
y_pred = gnb.predict(X_test)
```

# Confusion Matrix of the Classification Results

```python
confusion_matrix_Simulation = metrics.confusion_matrix(y_test,y_pred)
Simulation_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix_Simulation, display_labels = ["M", "B"])
Simulation_display.plot()
plt.show()

print("\nClassification Report:\n")
print(classification_report(y_test, y_pred, target_names=["M", "B"]))
```
